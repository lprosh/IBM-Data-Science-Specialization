## Проверка типов данных

Здесь мы проверяем типы данных, потому что некоторые из них могут храниться в виде строки, когда должны быть числовым значением, поэтому в этом случае нужно привести строку к целочисленному / вещественному формату. И только после отобразить данные с помощью графика. В данном случае данные уже находятся в необходимом формате.

## Переименование столбцов

Иногда имена столбцов запутаны для чтения, поэтому хорошей практикой будет их переименование. Такой подход улучшает читаемость набора данных.

```python
df = df.rename(columns={"HOURLYVISIBILITY": "Visibility", "HOURLYDRYBULBTEMPF": "DBT F", "HOURLYWETBULBTEMPF": "WBT F", 
                            "HOURLYDewPointTempF": "Dew Point Temp", "HOURLYRelativeHumidity" : "Relative Humidity", "HOURLYWindSpeed": "Wind Speed", 
                            "HOURLYStationPressure": "Station Pressure", "HOURLYSeaLevelPressure": "Sea Level Pressure", "HOURLYPrecip" : "Precip", 
                            "HOURLYAltimeterSetting" : "Altimeter Setting", "HOURLYWindDirectionSin" : "Wind Direction Sin", 
                            "HOURLYWindDirectionCos" : "Wind Direction Cos", "HOURLYPressureTendencyIncr" : "Pressure Tendency Incr", 
                            "HOURLYPressureTendencyDecr" : "Pressure Tendency Decr", "HOURLYPressureTendencyCons" : "Pressure Tendency Cons"})
df.head(5)
```

## Удаление повторяющихся строк
Часто огромный набор данных содержит некоторые повторяющиеся данные, которые могут вызывать беспокойство, поэтому далее приведён пример для удалениях всех повторяющихся значений из набора. Для данного набора дублей не найдено.

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("Число дублей: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

## Обнаружение Выбросов
Выброс — это точка или набор точек, которые отличаются от других точек. Иногда они могут быть очень высокими или очень низкими. Часто бывает хорошей идеей обнаружить и удалить выбросы. Потому что выбросы являются одной из основных причин, приводящих к получению менее точной модели. 

Далее применяется метод оценки IQR. Часто выбросы можно увидеть с помощью визуализаций, использующих прямоугольную диаграмму. Ниже показаны графики относительной влажности, статического давления и атмосферного давления. На некоторых графиках можно увидеть точки, которые выходят за рамки, они являются не чем иным, как выбросами. 

Техника поиска и удаления выбросов взята из учебника [Towards Data Science](https://towardsdatascience.com/ways-to-detect-and-remove-the-outliers-404d16608dba).

```python
sns.boxplot(x=df['Relative Humidity'])
```

```python
sns.boxplot(x=df['Station Pressure'])
```

```python
sns.boxplot(x=df['Sea Level Pressure'])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

## Тепловые карты

Тепловые карты — это тип графика, который необходим, когда нужно найти зависимые переменные. Одним из лучших способов найти взаимосвязь между объектами может быть использование тепловых карт.

```python
plt.figure(figsize=(20, 10))
c = df.corr()
sns.heatmap(c, cmap="BrBG", annot=True)
c
```
