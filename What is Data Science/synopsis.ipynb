# Неделя 1

### Определение Науки о данных

Наука о данных (Data Science) — это область изучения, обработки и анализа данных, а также использования данных для ответов на вопросы или выработки рекомендаций. Иными словами, наука о данных – это изучение больших объемов данных, которые могут выявить идеи, помогающие организациям принимать стратегические решения.

Поскольку наука о данных не является дисциплиной, традиционно преподаваемой в университетах, современные учёные по данным имеют разный опыт работы, например, в области инженерии, статистики и физики. Существует множество путей к карьере в области науки о данных; большинство, но не все, включают в себя немного математики, немного науки и большое любопытство к данным.

Новые учёные по данным должны быть любознательными, склонным к аргументации и уметь отстаивать собственную позицию.

Наука о данных считается самой привлекательной работой в 21 веке, где квалифицированным работникам платят высокую зарплату.

### Чем занимается учёный по данным?

Типичный рабочий день учёного по данным варьируется в зависимости от того, над каким типом проекта он работает.

Для извлечения информации из данных используется множество алгоритмов.

Использование сложных алгоритмов машинного обучения не всегда гарантирует достижение более высокой производительности. Иногда простой алгоритм, такой как метод k-ближайших соседей, может дать удовлетворительную производительность, сравнимую с той, которая достигается с помощью сложного алгоритма. Все зависит от данных. В любой области – и наука о данных не исключение – простое решение всегда предпочтительнее сложного, особенно если их производительность можно сравнить.

Доступ к алгоритмам, инструментам и данным через облако позволяет учёным оставаться в курсе последних событий и легко одновременно обращаться к данным, совместно работая над задачами.

## Большие данные и интеллектуальный анализ данных

### Основы больших данных

Нет единого определения Big Data, но есть определенные общие понятия в различных толкованиях, такие как скорость, объем, разнообразие, достоверность и ценность.

Пять V понятий Big Data:

- **Velocity (скорость)** — это скорость накопления данных. Данные генерируются очень быстро, и этот процесс никогда не останавливается. Технологии потокового вещания, облачные и серверные технологии могут обрабатывать информацию очень быстро.

- **Volume (объем)** означает масштаб данных, или прирост данных к существующему объему. Драйверами объема являются увеличение числа источников данных, рост разрешения сенсоров и масштабируемая инфраструктура. 

- **Variety** означает разнообразие данных. Структурированные данные прекрасно представимы в виде строк и столбцов, тогда как реляционные базы данных и неструктурированные данные, такие как твиты, посты в блогах, изображения, числа и видео, — не сведены к заранее заданному формату. Разнообразие также означает, что данные поступают из различных источников: от разных устройств, людей, процессов, — как внутренних, так и внешних по отношению к организациям. Драйверами являются мобильные технологии, социальные медиа, носимые устройства, геотехнологии, видео и многие другие. 

- **Veracity (достоверность)** — это вопрос качества и происхождения данных, их соответствие фактам и точность. Атрибуты достоверности — это согласованность, полнота, целостность и непротиворечивость*. [в оригинале "ambiguity" (двусмысленность) - прим.] К драйверам достоверности относятся стоимость и потребность в отслеживании. При таком большом количестве доступных данных возникает много споров о точности данных в цифровую эру. Соответствует ли информация действительности, или нет? 

- **Value (ценность)** — это наша способность и потребность превращать данные в ценность. Ценность — это не только прибыль. Это какие-то медицинские или социальные преимущества, а также удовлетворенность клиентов, сотрудников и личная удовлетворенность. 

По словам доктора Уайта, большинство компонентов науки о данных, таких как вероятность, статистика, линейная алгебра и программирование, существуют уже много десятилетий, но теперь у нас есть вычислительные возможности для их применения, объединения и разработки новых методов и алгоритмов обучения.

Hadoop и другие инструменты в сочетании с распределенными вычислительными мощностями используются для удовлетворения потребностей больших объёмов данных.

По словам доктора Уайта, его студентам, которые в основном являются начинающими учёными по данным, необходимо изучить множество инструментов, таких как Python, UNIX-команды, pandas и Jupyter notebook.

### Интеллектуальный анализ данных (Data Mining, DM)

#### Определение целей DM
Необходимо определить ключевые вопросы, на которые необходимо ответить, определить ожидаемый уровень точности и полезности результатов, полученных в результате DM. Компромисс между затратами и выгодами всегда играет важную роль в определении целей и масштабов DM. Уровень точности, ожидаемый от результатов, также влияет на затраты. Высокий уровень точности DM обойдётся дороже, и наоборот. Кроме того, за пределами определенного уровня точности можно не получить много пользы от DM. Таким образом, компромисс между затратами и выгодами для достижения желаемого уровня точности является важным фактором для целей DM.

#### Выбор данных
Результаты DM в значительной степени зависят от качества используемых данных. Иногда данные легко доступны для дальнейшей обработки. Например, розничные торговцы часто располагают большими базами данных о покупках и демографии клиентов. С другой стороны, данные могут быть недоступны для DM. В таких случаях нужно определить другие источники данных или даже спланировать новые инициативы по сбору данных, включая опросы. Тип данных, их размер и частота сбора напрямую влияют на стоимость DM. Поэтому определение правильного типа данных, необходимых для DM, которые могли бы ответить на вопросы при разумных затратах, имеет решающее значение.

#### Предобработка данных
Предварительная обработка данных является важным шагом в DM. Часто необработанные данные являются беспорядочными, содержащими ошибочные или нерелевантные данные. Кроме того, даже при наличии соответствующих данных иногда отсутствует информация. На этапе предварительной обработки определяются нерелевантные атрибуты данных и исключаются из дальнейшего рассмотрения. В то же время необходимо выявить ошибочные аспекты набора данных и пометить их как таковые.

Если бы данные отсутствовали случайным образом, было бы достаточно простого набора решений. Однако, когда данные отсутствуют систематически, следует определить влияние отсутствующих данных на результаты. Например, определенная подгруппа лиц в большом наборе данных, возможно, отказалась раскрывать свои доходы. Выводы, основанные на доходах отдельного лица в качестве исходных данных, исключили бы сведения о тех лицах, о доходах которых не сообщалось. Это привело бы к систематическим искажениям в анализе. Поэтому нужно заранее обдумать, следует ли исключить наблюдения или переменные, содержащие недостающие данные, из всего анализа или его частей.

#### Преобразование данных
После сохранения соответствующих атрибутов данных следующим шагом является определение соответствующего формата, в котором должны храниться данные. Важным соображением при DM является сокращение количества атрибутов, необходимых для объяснения явлений. Это может потребовать понижения размерности ([data reduction](https://wiki.loginom.ru/articles/data-reduction.html?ysclid=l5tee0f9v5452940523)); такие как метод главных компонент, могут уменьшить количество атрибутов без значительной потери информации. Кроме того, переменные, возможно, потребуется преобразовать, чтобы помочь объяснить изучаемое явление. Например, доход физического лица может быть зарегистрирован в наборе данных как доход от заработной платы; доход из других источников, таких как аренда недвижимости; выплаты поддержки от правительства и тому подобное. Агрегирование доходов из всех источников позволит получить репрезентативный показатель для индивидуального дохода.

Часто необходимо преобразовать переменные из одного типа в другой. Возможно, было бы разумно преобразовать непрерывную переменную для дохода в категориальную переменную, где каждая запись в базе данных идентифицируется как индивид с низким, средним и высоким доходом. Это могло бы помочь выявить нелинейности в лежащем в основе поведении.

#### Хранение данных
Преобразованные данные должны храниться в формате, который делает их удобными для интеллектуального анализа данных. Данные должны храниться в формате, который предоставляет учёному по данным неограниченные и немедленные права на чтение/запись. Во время DM создаются новые переменные, которые записываются обратно в исходную базу данных, поэтому схема хранения данных должна способствовать эффективному чтению и записи в базу данных. Также важно хранить данные на серверах или носителях, что обеспечивает безопасность данных, а также предотвращает ненужный поиск алгоритмом DM фрагментов данных, разбросанных на разных серверах или носителях. Безопасность и конфиденциальность данных должны быть главной заботой при хранении данных.

#### Интеллектуальный анализ данных
После того, как данные надлежащим образом обработаны, преобразованы и сохранены, они подвергаются DM. Этот шаг охватывает методы анализа данных, включая параметрические и непараметрические методы, а также алгоритмы машинного обучения. Хорошей отправной точкой для DM является визуализация данных. Многомерные представления данных с использованием расширенных возможностей построения графиков программного обеспечения DM очень полезны для получения предварительного понимания тенденций, скрытых в наборе данных.

#### Оценка результатов DM
После того как результаты были извлечены из DM, проводится формальная оценка результатов. Формальная оценка может включать тестирование прогностических возможностей моделей на наблюдаемых данных, чтобы увидеть, насколько эффективными и действенными оказались алгоритмы при воспроизведении данных. Это известно как "прогноз в выборке". Кроме того, результаты передаются ключевым заинтересованным сторонам для получения обратной связи, которая затем включается в последующие итерации DM для улучшения процесса.

DM и оценка результатов становятся итеративным процессом, так что аналитики используют более совершенные и усовершенствованные алгоритмы для повышения качества результатов, полученных в свете обратной связи, полученной от ключевых заинтересованных сторон.

## Глубокое обучение и машинное обучение

### В чём разница?

Термин "большие данные" относится к наборам данных настолько массивным, так быстро накапливаемым и настолько разнообразным, что они не поддаются традиционным методам анализа, которые применяются к реляционным базам данных. Достижение огромных вычислительных мощностей распределённых сетей и параллельная ему разработка новых инструментов и методов анализа данных означает, что теперь организации теперь имеют мощности для анализа этих огромных массивов данных. 

Data Mining — это процесс автоматического поиска и анализа данных, выявление скрытых прежде закономерностей. Он включает в себя предварительную обработку данных, чтобы подготовить их и преобразовать в нужный формат. Как только это будет сделано, происходит извлечение идей и закономерностей с помощью различных инструментов и методов, начиная от простых средств визуализации данных, и заканчивая машинным обучением и статистическими моделями. Машинное обучение является подмножеством ИИ, использующим компьютерные алгоритмы для анализа данных и принятия интеллектуальных решений на основе того, чему они были обучены без непосредственного программирования.

- Машинное обучение — это то, что позволяет машинам решать задачи самостоятельно, делая точные прогнозы на основе предоставленных данных. 

- Глубокое обучение — это тип машинного обучения, который имитирует процесс принятия решений человеком с использованием нейронных сетей.

- Нейронная сеть в ИИ — это набор мелких вычислительных блоков, называемых нейронами. Они принимают данные на вход и со временем учатся принимать решения.

Машинное обучение имеет множество применений, от рекомендательных систем, которые предоставляют клиентам релевантный выбор на коммерческих веб-сайтах, до детального анализа финансовых рынков.

### Нейросети и глубокое обучение

Нейронная сеть — это попытка использовать компьютер, компьютерную программу, имитирующую то, как наш мозг использует нейроны и синапсы, обрабатывая информацию.

Нейронные сети существуют уже несколько десятилетий, но на некоторое время от них отказались, потому что они были очень дорогостоящими с точки зрения вычислений.

Варианты использования глубокого обучения:

- Классификация изображений в большом масштабе.

- Распознавание речи.

Варианты применения машинного обучения:

- Рекомендательные системы. Например, Netflix использует машинное обучение, чтобы рекомендовать вам фильмы на основе фильмов, которые вы уже посмотрели и которые вам понравились или не понравились.

- Выявление случаев мошенничества.
